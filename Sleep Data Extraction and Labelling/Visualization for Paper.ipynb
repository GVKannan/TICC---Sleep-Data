```python
#Iterations vs Time vs Beta:
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
iters300 = [1,5,10]
BIC300 = [45781679.261003055, 6452830.5600768188, 45718.6419477]
lnBIC300 = np.log10(BIC300)
time300 = [25.44, 89.45, 169.091]
ax1.plot(iters300,BIC300,'bo')
ax1.plot(iters300,BIC300)
ax1.set_xlabel('Iterations')
ax1.set_ylabel('BIC')


ax2.plot(iters300,time300,'bo')
ax2.plot(iters300,time300)
ax2.set_xlabel('Iterations')
ax2.set_ylabel('Time')
#savefig("Beta = 300; Iters vs BIC & Iters vs Time")
```
